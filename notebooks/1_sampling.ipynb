# Credit Risk Analysis - Data Sampling

This notebook creates a balanced and representative sample from the large Lending Club dataset (1.19GB) for efficient analysis and modeling.

```python
import pandas as pd
print("Working!")
```

```python
print("Loading dataset...")

df = pd.read_csv("../data/loan.csv")

print("Dataset loaded successfully!")
```

```python
df.shape

```

```python
df.columns
```

```python
df['loan_status'].value_counts()
```

```python
df = df[df['loan_status'].isin(['Fully Paid', 'Charged Off'])]
```

```python
df['loan_status'].value_counts(normalize=True)
```

```python
print("Creating stratified sample...")

good = df[df['loan_status'] == 'Fully Paid']
bad = df[df['loan_status'] == 'Charged Off']

df_sample = pd.concat([
    good.sample(n=100000, random_state=42),
    bad.sample(n=100000, random_state=42)
])

print("Sample created!")
```

```python
df_sample = df_sample.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
df_sample['loan_status'].value_counts()
```

```python
df_sample.to_csv("../data/loan_sample.csv", index=False)

print("Sample saved successfully!")
```
